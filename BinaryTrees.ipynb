```python
import pandas as pd

class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None

root = TreeNode(0)
left = TreeNode(1)
right = TreeNode(2)
root.left = left
root.right = right

print(root.left.val)
print(root.right.val)

```
